```python
# ============================================================
# 01_eda.ipynb â€” Exploratory Data Analysis (EDA)
# Dataset: job_change
# Goal: understand data structure, target balance, missingness,
#       feature types, and basic distributions (no modeling here).
# ============================================================

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option("display.max_columns", None)
pd.set_option("display.width", 120)

# ----------------------------
# 0) (Optional) Repo sanity check
# ----------------------------
print("Current working directory:", os.getcwd())
print("Top-level files/folders:", os.listdir(".")[:20])

# ----------------------------
# 1) Load data
# ----------------------------
TRAIN_PATH = "data/job_change_train.csv"
TEST_PATH  = "data/job_change_test.csv"
TARGET = "willing_to_change_job"
ID_COL = "id"

train = pd.read_csv(TRAIN_PATH)
test  = pd.read_csv(TEST_PATH)

print("\nShapes")
print("Train:", train.shape)
print("Test :", test.shape)

print("\nTrain columns:", list(train.columns))
print("Test columns :", list(test.columns))

# Quick preview
display(train.head())
display(test.head())

# ----------------------------
# 2) Basic checks
# ----------------------------
print("\nTarget in train?", TARGET in train.columns)
print("Target in test ?", TARGET in test.columns)

# ID uniqueness
if ID_COL in train.columns:
    print("\nID uniqueness check (train):")
    print("Unique IDs:", train[ID_COL].nunique(), "| Rows:", train.shape[0])

# Duplicated rows (optional)
print("\nDuplicated rows in train:", train.duplicated().sum())
print("Duplicated rows in test :", test.duplicated().sum())

# ----------------------------
# 3) Target distribution (class balance)
# ----------------------------
print("\nTarget distribution (counts):")
print(train[TARGET].value_counts(dropna=False))

print("\nTarget distribution (shares):")
print(train[TARGET].value_counts(normalize=True, dropna=False).rename("share").to_frame())

# Plot target distribution
plt.figure()
train[TARGET].value_counts().plot(kind="bar")
plt.title("Target distribution: willing_to_change_job")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()

# ----------------------------
# 4) Missing values
# ----------------------------
missing_train = (train.isna().mean() * 100).sort_values(ascending=False)
missing_test  = (test.isna().mean() * 100).sort_values(ascending=False)

print("\nTop missing columns (train):")
display(missing_train.head(15).to_frame("missing_%"))

print("\nTop missing columns (test):")
display(missing_test.head(15).to_frame("missing_%"))

# Columns with any missingness
cols_with_missing = missing_train[missing_train > 0].index.tolist()
print("\nColumns with missing values (train):", cols_with_missing)

# ----------------------------
# 5) Feature types (categorical vs numerical)
# ----------------------------
# We'll define features as all train columns except target
feature_cols = [c for c in train.columns if c != TARGET]

X_train = train[feature_cols].copy()

# Categoricals: object (strings)
cat_cols = X_train.select_dtypes(include=["object"]).columns.tolist()

# Numerics: int/float
num_cols = [c for c in X_train.columns if c not in cat_cols]

print("\nCategorical columns (object):", cat_cols)
print("\nNumeric columns:", num_cols)

print("\nDtypes summary:")
display(train[feature_cols].dtypes.to_frame("dtype").T)

# ----------------------------
# 6) Quick summaries of numeric features
# ----------------------------
if len(num_cols) > 0:
    print("\nNumeric summary (train):")
    display(train[num_cols].describe().T)

# ----------------------------
# 7) Frequency tables for key categorical features
# ----------------------------
# show top categories for each categorical column (train)
MAX_CATS_TO_SHOW = 10
for col in cat_cols:
    print(f"\nCategorical feature: {col}")
    vc = train[col].value_counts(dropna=False).head(MAX_CATS_TO_SHOW)
    display(vc.to_frame("count"))

# ----------------------------
# 8) Check for special "range-coded" variables
#     (e.g. values like '>4', '>20' which should be treated as categorical)
# ----------------------------
# These often appear in years_since_job_change / years_of_experience (as seen in your preview).
range_like_cols = []
for col in feature_cols:
    if train[col].dtype == "object":
        # If column has '>' symbols, it's likely range-coded
        if train[col].astype(str).str.contains(">", regex=False).any():
            range_like_cols.append(col)

print("\nColumns containing '>' (likely range-coded categorical):", range_like_cols)

# Show unique values (limited) for those columns
for col in range_like_cols:
    print(f"\nUnique values (sample) for: {col}")
    uniques = pd.Series(train[col].unique()).astype(str)
    display(uniques.head(30))

# ----------------------------
# 9) Simple plots for a few numeric variables (optional, keep it light)
# ----------------------------
# Age distribution
if "age" in train.columns:
    plt.figure()
    train["age"].hist(bins=30)
    plt.title("Age distribution")
    plt.xlabel("Age")
    plt.ylabel("Count")
    plt.show()

# Relative wage distribution
if "relative_wage" in train.columns:
    plt.figure()
    train["relative_wage"].hist(bins=30)
    plt.title("Relative wage distribution")
    plt.xlabel("Relative wage")
    plt.ylabel("Count")
    plt.show()

# Hours of training distribution
if "hours_of_training" in train.columns:
    plt.figure()
    train["hours_of_training"].hist(bins=30)
    plt.title("Hours of training distribution")
    plt.xlabel("Hours of training")
    plt.ylabel("Count")
    plt.show()

# ----------------------------
# 10) Quick relationship checks (optional, still EDA)
#     For example: mean age by target, and target rate by category
# ----------------------------
# Numeric vs target (means)
if "age" in train.columns:
    print("\nMean age by target:")
    display(train.groupby(TARGET)["age"].mean().to_frame("mean_age"))

# Target rate by a few categorical columns (top categories)
def target_rate_table(df, col, target=TARGET, top_n=10):
    tmp = df[[col, target]].copy()
    # Convert target to numeric just in case
    tmp[target] = pd.to_numeric(tmp[target], errors="coerce")
    counts = tmp[col].value_counts(dropna=False).head(top_n).index
    tmp = tmp[tmp[col].isin(counts)]
    rate = tmp.groupby(col)[target].mean().sort_values(ascending=False)
    return rate.to_frame("target_rate")

for col in ["gender", "education", "field_of_studies", "is_studying", "type_of_company", "size_of_company"]:
    if col in train.columns and col in cat_cols:
        print(f"\nTarget rate by {col} (top categories):")
        display(target_rate_table(train, col, top_n=10))

```
